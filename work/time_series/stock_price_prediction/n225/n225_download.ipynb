```python
import torch
import torch.nn as nn

class CNN1d(nn.Module):
    def __init__(
            self, 
            seq_len, 
            input_channels, 
            n_features, 
            n_filters,
            droprate: float = 0.1,
        ):
        super().__init__()
        conv_output_size = 10 # calc
        self.l1 = nn.Covn1d(input_channels, n_filters[0], kernel_size=n_features[0])
        self.a1 = nn.ReLU()
        self.l2 = nn.MaxPool1d(kernel_size=2)
        self.l3 = nn.Conv1d(n_features[2], n_features[3], kernel_size=2)
        self.a3 = nn.ReLU()
        self.l4 = nn.Flatten()
        self.l5 = nn.Dropout(droprate)
        self.l6 = nn.Dense(conv_output_size, 1)
        self.layers = [
            self.l1, self.a1,
            self.l2, 
            self.l3, self.a3,
            self.l4,
            self.l5,
            self.l6
        ]
        
    def forward(self, x: torch.Tensor):
        for layer in self.layers:
            x = layer(x)
        
        return x
```
